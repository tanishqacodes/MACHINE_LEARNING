```python
from sklearn.datasets import make_classification
# helps to create a data set of classification(catorigcal kind)
```

```python
make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=1,n_redundant=0)
```

```python
#x is the independent feature 
#y is the dependent feature 
#nsample means =1000 number of data fs
X,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
X
```

```python
y
```

```python
import pandas as pd
```

```python
df1  = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(final_df[['f1','f2']] , final_df['target'])
```
